```python
data.dropna(axis=1, how='all', inplace=True)
```

```python
st=data['search_type']
st.unique()
# replace NAN values
data.fillna({
    'search_type':'No Search'
}, inplace=True)
```

```python
data['stop_date']=pd.to_datetime(data['stop_date'], format='%Y-%m-%d')
data['stop_time']=pd.to_datetime(data['stop_time'], format='%H:%M:%S')
```

```python
data.drop(['driver_age_raw'],axis=1,inplace=True)
```

```python
# connect to MySQL
connection = pymysql.connect(
    host="localhost",
    user="root",
    password="Mathi@1903",
    database="policedb"
)

cursor = connection.cursor()
print("✅ Connected successfully")
```

```python
cursor.execute("SHOW DATABASES")
for db in cursor.fetchall():
    print(db)
```

```python
cursor.execute("CREATE DATABASE IF NOT EXISTS policedb")
cursor.execute("USE policedb")
connection.commit()
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS logs (
    id INT AUTO_INCREMENT PRIMARY KEY,
    stop_date DATE,
    stop_time TIME,
    country_name VARCHAR(50),
    driver_gender VARCHAR(10),
    driver_age INT,
    driver_race VARCHAR(50),
    violation_raw VARCHAR(100),
    violation VARCHAR(100),
    search_conducted BOOLEAN,
    search_type VARCHAR(100),
    stop_outcome VARCHAR(100),
    is_arrested BOOLEAN,
    stop_duration VARCHAR(50),
    drugs_related_stop BOOLEAN,
    vehicle_number VARCHAR(50)
)
""")
```

```python
connection.commit()
print("✅Table logs created successfully")
```

```python
data_tuples = [
    (
        row['stop_date'], row['stop_time'], row['country_name'], row['driver_gender'], row['driver_age'],
        row['driver_race'], row['violation_raw'], row['violation'], row['search_conducted'],
        row['search_type'], row['stop_outcome'], row['is_arrested'], row['stop_duration'],
        row['drugs_related_stop'], row['vehicle_number']
    )
    for _, row in data.iterrows()
]

cursor.executemany("""
    INSERT INTO logs (
        stop_date, stop_time, country_name, driver_gender, driver_age,
        driver_race, violation_raw, violation, search_conducted, search_type,
        stop_outcome, is_arrested, stop_duration, drugs_related_stop, vehicle_number
    ) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
""", data_tuples)

connection.commit()
```

```python
cursor.close()
connection.close()
print("✅Data inserted into SECURECHECK.logs using iterrows()")
```
